```python
def total_skills(soup):
    total_Skills=[]
    for i in range(0,len(soup.find_all('span', {'itemprop': 'programmingLanguage'}))-1):
        #for j in range(i+1,len(soup.find_all('span', {'itemprop': 'programmingLanguage'}))):    
            #if(soup.find_all('span', {'itemprop': 'programmingLanguage'})[j] != soup.find_all('span', {'itemprop': 'programmingLanguage'})[j]):

                total_Skills.append(soup.find_all('span', {'itemprop': 'programmingLanguage'})[i].get_text().strip())
    return total_Skills    
```

```python
def unique_skills(soup):
    unique_Skills=[]
    total_Skills=total_skills(soup) 
    for i in total_Skills:
        if i not in unique_Skills:    
            unique_Skills.append(i)           
    return unique_Skills  
```

```python
def pourcentage_skills(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.text,'html.parser')
    total_Skills=total_skills(soup)
    unique_Skills=unique_skills(soup)
    Skills_pour=dict()
    oc=[]
    nn=[]
    info=dict()
    for i in unique_Skills:
        k=0
        for j in total_Skills:        
            if i ==j:
                k=k+1
        nn.append(i)
        oc.append([round(k*100/len(total_Skills),0)])    
    for j in range(len(nn)):
        info[nn[j]+"_approved"]=oc[j]
    return info 
```
